importing the dependencies

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets as data
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the boston house price dataset


```python
house_price_datasets = data.load_boston()
print(house_price_datasets)
```

```python
#loading the datasets to pandas data frame
house_price_dataframe = pd.DataFrame(house_price_datasets.data,columns=house_price_datasets.feature_names)
```

```python
#print first 5 rows of our data frame
house_price_dataframe.head()
```

```python
#add the target (price ) column to the dataframe
house_price_dataframe['price'] = house_price_datasets.target
```

```python
house_price_dataframe.head()
```

```python
#checking the number of rows and column in the datafame
house_price_dataframe.shape
```

```python
#cheak for missing values
house_price_dataframe.isnull().sum()
```

```python
#stastitical measures of the datasets
house_price_dataframe.describe()
```

Understanding the correlation between various features in the datasets

Positive correlation,
Negative correlation




```python
correlation = house_price_dataframe.corr()
```

```python
#constructing a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True, fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

Splitting the data and target


```python
X = house_price_dataframe.drop(['price'],axis=1)
Y=house_price_dataframe['price']
print(X)
print(Y)
```

Splitting the data into training and test data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model training

XGBoost Regressor

```python
#more of desition tree model made up of many models
#loading the model
model = XGBRegressor()
```

```python
#training the model with X_train
model.fit(X_train,Y_train)
```

Evaluation (how well our model is predicting)

Prediction on training data

```python
#accuracy for prediction of training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
#R squared error
score_1 = metrics.r2_score(Y_train,training_data_prediction)
#Mean absolute error
score_2 = metrics.mean_absolute_error(Y_train,training_data_prediction)

print('R Square error : ',score_1)
print('Mean absolute error : ',score_2)
#less the value means the more accurate our value
```

visualizing the actual prices and predicted prices

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Prices')
plt.title('Actual Price vs Predicted Price')
plt.show()
```

Predition on test data

```python
#accuracy for prediction of test data
test_data_prediction = model.predict(X_test)
```

```python
#R squared error
score_1 = metrics.r2_score(Y_test,test_data_prediction)
#Mean absolute error
score_2 = metrics.mean_absolute_error(Y_test,test_data_prediction)

print('R Square error : ',score_1)
print('Mean absolute error : ',score_2)
#less the value means the more accurate our value
```
